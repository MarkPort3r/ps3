```python
# Problem 1
##read through water level data and report highest water level and the date and time that it was observed.

import csv

max_value = float('-inf')  # Initialize with negative infinity to track the max
max_row = None  # To store the row with the maximum value

with open('/blue/bsc4452/markporter/PS3/ps3/CO-OPS_8727520_wl.csv', 'r') as file:
    reader = csv.reader(file) #CSV reader reads each row in as a list we can query
    next(reader)  # Skip the header

    for row in reader:
        try:
            value = float(row[3])  # Convert the 4th field (Preliminary (ft)) to float
            if value > max_value:
                max_value = value
                max_row = row
        except ValueError:
            # Handle cases where the 4th field is not a number (e.g., "-")
            continue

if max_row:
    print(f"High Water Level: {max_value} ft")
    print(f"Date: {max_row[0]}, Time: {max_row[1]}")
```

```python
# Problem 2
##read through water level data and report lowest, highest and average water level and the date and time that it was observed.

import csv

max_value = float('-inf')  # Initialize with negative infinity to track the max
min_value = float('inf')
sum_values = 0.0
count = 0
max_row = None  # To store the row with the maximum value
min_row = None

with open('/blue/bsc4452/markporter/PS3/ps3/CO-OPS_8727520_wl.csv', 'r') as file:
    reader = csv.reader(file) #CSV reader reads each row in as a list we can query
    next(reader)  # Skip the header

    for row in reader:
        try:
            value = float(row[3])  # Convert the 4th field (Preliminary (ft)) to float
            if value > max_value:
                max_value = value
                max_row = row
            if value < min_value:
                min_value = value
                min_row = row
                
            sum_values += value
            count += 1
                
        except ValueError:
            # Handle cases where the 4th field is not a number (e.g., "-")
            continue

if count > 0:
    avg_value = sum_values / count
else:
    avg_value = None

#results
if max_row:
    print(f"High Water Level: {max_value} ft on {max_row[0]} at {max_row[1]}")
if min_row:
    print(f"Low Water Level: {min_value} ft on {min_row[0]} at {min_row[1]}")
if avg_value is not None:
    print(f"Average Water Level: {avg_value:.2f} ft")
else:
    print("No valid data to calculate average water level")
```

```python
# Problem 3
##calculate fastest rise in water level between lines.

import csv

prev_value = None
max_diff_row = None
previous_row = None  # To store the previous row for comparison
max_diff = 0

with open('/blue/bsc4452/markporter/PS3/ps3/CO-OPS_8727520_wl.csv', 'r') as file:
    reader = csv.reader(file)  # CSV reader reads each row in as a list we can query
    next(reader)  # Skip the header

    for row in reader:
        try:
            value = float(row[3])  # Convert the 4th field (Preliminary (ft)) to float
            
            if prev_value is not None:
                diff = (value - prev_value)
                if diff > max_diff:
                    max_diff = diff
                    max_diff_row = row  # Store the current row where the largest difference occurs
                    # Store the previous row before updating it for the next iteration
                    previous_row = prev_row  

            # Update prev_value and prev_row to the current row's values for the next iteration
            prev_value = value
            prev_row = row  # Store the current row as the previous row for the next comparison
            
        except ValueError:
            # Handle cases where the 4th field is not a number (e.g., "-")
            continue

# Print the largest difference
if max_diff_row and previous_row:
    print(f"Fastest rise in water level over 6 minutes: {max_diff} ft occurred between {previous_row[1]} and {max_diff_row[1]} on {max_diff_row[0]}.")
    print(f"The water level went from {previous_row[3]} ft to {max_diff_row[3]} ft.")
```

```python
# Problem 4
##Read file as if it were a live update and thro errots for increase of 0.25, water level over 5 ft or no recording.

import csv

prev_value = None
prev_row = None  # To store the previous row for comparison

with open('/blue/bsc4452/markporter/PS3/ps3/CO-OPS_8727520_wl.csv', 'r') as file:
    reader = csv.reader(file)  # CSV reader reads each row in as a list we can query
    next(reader)  # Skip the header

    for row in reader:
        try:
            value = float(row[3])  # Convert the 4th field (Preliminary (ft)) to float
            
            # Check if value > 5 and print a warning
            if value > 5:
                print(f"WARNING: Water level {value} ft exceeds 5 ft on {row[0]} at {row[1]}")

            # Check if there's a previous value to compare with
            if prev_value is not None:
                # Check if the difference between current and previous values is > 0.25
                diff = abs(value - prev_value)
                if diff > 0.25:
                    print(f"WARNING: Water level changed by more than 0.25 ft ({diff} ft) between {prev_row[0]} {prev_row[1]} and {row[0]} {row[1]}")
            
            # Update prev_value and prev_row for the next iteration
            prev_value = value
            prev_row = row  # Store the current row as the previous row for the next comparison

        except ValueError:
            # Handle cases where the 4th field is not a valid float (e.g., "-")
            print(f"WARNING: Non-numeric water level value '{row[3]}' on {row[0]} at {row[1]}")
            continue
```
